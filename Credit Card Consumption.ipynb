```python
def missing_var(x):
    if ((x.dtype=="float") or (x.dtype == "int64")):
        x = x.fillna(x.median())
    elif ((x.dtype=="object") or (x.dtype == "O")):
        x = x.fillna(x.mode()[0])
    else:
        x
    return x
```

```python
existing.drop(columns={'personal_loan_active','vehicle_loan_active','vehicle_loan_closed'},inplace=True)
```
